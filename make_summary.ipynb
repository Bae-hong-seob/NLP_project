# 1. 데이터 로드
  - 제출용 예시 output은 가장 row가 적은 4번 카테고리 dataset에 대해 진행.
  - 타 dataset(3만개 데이터)은 summary와 sentiment analysis 진행 시 colab gpu 사용시 5시간가량 소요

```python
import pandas as pd
import ast
import warnings
warnings.filterwarnings(action='ignore')

file_path = './splited_dataset/'
file_name = 'sample_total_df_04_10.csv'

total_df = pd.read_csv(file_path + file_name) # 02: beaty and sopa
total_df
```

```python
try:
    total_df = total_df.drop('Unnamed: 0', axis=1)
except:
    print('No Index Column')

total_df
```

# 2. review 데이터 감성분석 모델 구축

## 2.1. review 데이터 준비

```python
reviews = list(total_df['text'])
len(reviews)
```

```python
#num_samples = 100
num_samples = len(reviews)
```

```python
# 코드 실행시 example_reviews 데이터로 우선 진행
example_reviews = reviews[:num_samples]
example_reviews
```

## 2.2. 텍스트 요약
- bart-large-cnn 모델 사용

```python
# !pip install transformers
```

```python
from transformers import pipeline

#GPU 사용
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=0)

# Set the desired maximum length for the summary
max_summary_length = 512
```

```python
import pandas as pd

# 배치 사이즈 정의 : 요약 생성 시간이 꽤 걸리므로 배치 단위마다 저장
batch_size = 100
start_index = 0
end_index = len(example_reviews)

save_df = pd.DataFrame()
for i in range(start_index, end_index, batch_size):
    print(i)

    batch = example_reviews[i:i+batch_size]
    batch_summaries = summarizer(batch, max_length=max_summary_length, min_length=10, do_sample=False)

    batch_df = total_df[i:i+batch_size].copy()

    batch_summaries = [batch_summary['summary_text'] for batch_summary in batch_summaries]
    batch_df['summary'] = batch_summaries

    save_df = pd.concat([save_df, batch_df])
    save_df.to_csv(file_path + 'summary_'+str(start_index)+'_'+ file_name, index=False)

    #break
```
